## Purchasing Analysis (Total)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchasing_analysis_df = purchases_df.loc[:, ["Item ID", "Price", "Purchase ID"]]
# purchasing_analysis_df

# unique_items
unique_items = len(purchases_df["Item ID"].value_counts())
purchasing_analysis_df["Number of Unique Items"] = unique_items
# purchasing_analysis_df

# average_price and limit to two decimal places
average_price = purchases_df["Price"].mean()
purchasing_analysis_df["Average Price"] = average_price
# purchasing_analysis_df

# number_of_purchases
purchase_number = len(purchases_df["Purchase ID"].value_counts())
purchasing_analysis_df["Number of Purchases"] = purchase_number
# purchasing_analysis_df

# total_revenue
total_revenue = purchases_df["Price"].sum()
purchasing_analysis_df["Total Revenue"] = total_revenue
# purchasing_analysis_df

#reformat average price column
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format) 
# purchasing_analysis_df

#create summary dataframe
pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price], "Number of Purchases":[purchase_number], "Total Revenue":[total_revenue]})

#reformat total revenue column
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format) 
# purchasing_analysis_df

purchasing_analysis_df = purchasing_analysis_df.loc[:, ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish the bins and labels
bins = [0,9,14,19,24,29,34,39,100]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Add new column Age Group 
purchases_df["Age Group"] = pd.cut(purchases_df["Age"], bins, labels=group_labels)

# group by SN to avoid duplicate players
grouped_SN = purchases_df.groupby("SN").min()
grouped_age_df = pd.DataFrame(grouped_SN["Age Group"].unique().value_counts())

grouped_age_df = pd.DataFrame(grouped_SN.groupby("Age Group")['Age'].count())
grouped_age_df.columns=['Total Count']

grouped_age_df["Percentage"] = grouped_age_df["Total Count"]/player_count*100
grouped_age_df["Percentage"] = grouped_age_df["Percentage"].map("{:,.2f}".format)
grouped_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Establish the bins and labels
bins = [0,9,14,19,24,29,34,39,100]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Create Age Group column
purchases_df["Age Group"] = pd.cut(purchases_df["Age"], bins, labels=group_labels)

# Calculations
grouped_age_group = purchases_df.groupby(['Age Group'])

age_purchase_count = grouped_age_group["Purchase ID"].count()
avg_purchase_price = grouped_age_group["Price"].mean()
tot_purchase_value = grouped_age_group["Price"].sum()

# Create new Dataframe 
purchase_analysis = pd.DataFrame({'Purchase Count':age_purchase_count,'Average Purchase Price':avg_purchase_price,'Total Purchase Value':tot_purchase_value})

# Change groupby
avg_purchase_total_person = purchases_df.groupby(['SN']).agg({'Purchase ID':'min','Age':'min','Item ID':'min','Price':'sum','Age Group':'min'})

avg_purchase_total_person["Age Group"] = pd.cut(avg_purchase_total_person["Age"],bins,labels = group_labels)
avg_purchase_total_person_group = avg_purchase_total_person.groupby(['Age Group'])

# Add "Average Purchase Total per Person by Age Group" column 
purchase_analysis["Avg Total Purchase per Person"] = avg_purchase_total_person_group["Price"].mean()

# Format the Values
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis["Avg Total Purchase per Person"] = purchase_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

purchase_analysis
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Determine the most popular items
pop_items_group = purchases_df.groupby("Item ID").count().sort_values(by="Age",ascending = False).head(5)

# Get the rows of the most popular items
pop_items = purchases_df[purchases_df["Item ID"].isin(pop_items_group.index)]

# Calculations
pop_items_grp_total_pur_value = pd.DataFrame(pop_items.groupby("Item ID")["Price"].sum()).sort_values("Item ID")
pop_items_grp_total_pur_value = pop_items_grp_total_pur_value.rename(columns = {'Price':'Total Purchase Value'})

pop_items_grp_name = pd.DataFrame(pop_items.groupby("Item ID")["Item Name"].min()).sort_values("Item ID") #use min to select any same value

pop_items_pur_count = pd.DataFrame(pop_items.groupby("Item ID")["Item Name"].count()).sort_values("Item ID")
pop_items_pur_count = pop_items_pur_count.rename(columns = {'Item Name':'Purchase Count'})

pop_items_item_price = pd.DataFrame(pop_items.groupby("Item ID")["Price"].min()).sort_values("Item ID")


# Merge the DataFrames
merged_pd = pd.concat([pop_items_item_price,
           pop_items_grp_name,
           pop_items_pur_count,
           pop_items_grp_total_pur_value],axis=1)


# Format the Values
merged_pd["Price"] = merged_pd["Price"].map("${:.2f}".format)
merged_pd["Total Purchase Value"] = merged_pd["Total Purchase Value"].map("${:.2f}".format)


merged_pd
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Determine the most profitable items
profit_items_group = purchases_df.groupby("Item ID").sum().sort_values(by="Price",ascending = False).head(5)

# Get the rows of the most profitable items
prof_items = purchases_df[purchases_df["Item ID"].isin(profit_items_group.index)]

# Calculations
prof_items_name = pd.DataFrame(prof_items.groupby("Item ID")["Item Name"].min()).sort_values("Item ID")
prof_items_pur_count = pd.DataFrame(prof_items.groupby("Item ID")["Item Name"].count()).sort_values("Item ID")
prof_items_price = pd.DataFrame(prof_items.groupby("Item ID")["Price"].min()).sort_values("Item ID")
prof_items_tot_pur_value = pd.DataFrame(prof_items.groupby("Item ID")["Price"].sum()).sort_values("Item ID")

# Merge the DataFrames
most_prof_items = pd.concat([prof_items_name, prof_items_pur_count, prof_items_price, prof_items_tot_pur_value],axis=1)

most_prof_items.columns = ["Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]
most_prof_items.sort_values("Total Purchase Value", ascending = False)


# Format Item Price and Total Purchase Value
most_prof_items["Item Price"] = most_prof_items["Item Price"].map("${:.2f}".format)
most_prof_items["Total Purchase Value"] = most_prof_items["Total Purchase Value"].map("${:.2f}".format)


most_prof_items
```
